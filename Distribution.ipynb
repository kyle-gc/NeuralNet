```
import numpy as np
import random
import seaborn as sns

X = np.random.normal(40, 5, 10000)
print(X.shape)
```

```
sns.displot(X, kind="kde", fill=True)
```

```
SampleSize = 20

class Sample:
    def RS(X):
      return np.random.choice(X, SampleSize, replace=False)
```

```
X1 = Sample.RS(X)
X2 = Sample.RS(X)
X3 = Sample.RS(X)

XS = np.column_stack((X1, X2, X3))

XM1 = np.mean(X1)
XM2 = np.mean(X2)
XM3 = np.mean(X3)

sns.displot(XS, kind="kde", fill=True)
print('Mean (Sample 1) = ', XM1, 'Mean (Sample 2) = ', XM2, 'Mean (Sample 3) =', XM3)
```

```
Count = 0
Max_Count = 1000
Ave_Count = 0
Samp_Dist_Mean = []

while True:
  Outcome = Sample.RS(X)
  Average = np.mean(Outcome)
  Samp_Dist_Mean.append(Average)

  if Average >= 45:
    print(Average)
    Ave_Count +=1


  Count +=1

  if Count >= Max_Count:
    break

print(Ave_Count)

sns.displot(Samp_Dist_Mean, kind="kde", fill=True)
```
